```python
import os
import wfdb
import numpy as np

# Get the list of all files in the current directory
all_files = os.listdir('.')

# Filter for files that have the .dat extension (indicating ECG data files)
dat_files = [f for f in all_files if f.endswith('.dat')]

record_names = [os.path.splitext(f)[0] for f in dat_files]

for record_name in record_names:

    record = wfdb.rdrecord(record_name)  # Reads corresponding .dat and .hea files
    annotation = wfdb.rdann(record_name, 'atr')  # Reads .atr file


    ecg_signal = record.p_signal[:, 0]
    sampling_rate = record.fs

    # Display basic information about the record
    print(f"Loaded record: {record_name}")
    print(f"Signal shape: {record.p_signal.shape}")
    print(f"Sampling rate: {sampling_rate}")
    print(f"Annotations (sample indices): {annotation.sample}")
    print(f"Annotations (symbols): {annotation.symbol}")

```

```python
from scipy import signal

# Bandpass filter:
nyquist_freq = 0.5 * sampling_rate
low_cutoff = 0.5 / nyquist_freq
high_cutoff = 40 / nyquist_freq

b, a = signal.butter(1, [low_cutoff, high_cutoff], btype='band')
filtered_signal = signal.filtfilt(b, a, ecg_signal)
```

```python
# R-peak locations in samples (indices of R-peaks)
r_peaks = annotation.sample

# Define a window around each R-peak to extract the individual heartbeat
# E.g., 100 samples before and 200 samples after the R-peak (adjust as needed)
window_before = int(0.1 * sampling_rate)  # 100 ms before
window_after = int(0.2 * sampling_rate)   # 200 ms after

# Extract heartbeat segments
heartbeats = []
for r_peak in r_peaks:
    start = max(0, r_peak - window_before)
    end = min(len(filtered_signal), r_peak + window_after)
    heartbeat = filtered_signal[start:end]
    heartbeats.append(heartbeat)
```

```python
# Calculate RR intervals (difference between consecutive R-peak locations)
rr_intervals = np.diff(r_peaks) / sampling_rate  # Convert to seconds

# Example time-domain features for each heartbeat
time_features = []
for heartbeat in heartbeats:
    # Feature: Mean and standard deviation of the heartbeat
    mean_heartbeat = np.mean(heartbeat)
    std_heartbeat = np.std(heartbeat)

    # Combine features into a dictionary or list
    features = [mean_heartbeat, std_heartbeat]
    time_features.append(features)
```

```python
from scipy.fft import fft

frequency_features = []
for heartbeat in heartbeats:
    # Compute FFT and get power spectrum
    fft_values = np.abs(fft(heartbeat))
    power_spectrum = np.square(fft_values)

    # Feature: Sum of power in different frequency bands
    total_power = np.sum(power_spectrum)
    low_freq_power = np.sum(power_spectrum[:int(0.04 * len(power_spectrum))])
    high_freq_power = np.sum(power_spectrum[int(0.15 * len(power_spectrum)):int(0.5 * len(power_spectrum))])

    # Combine features into a dictionary or list
    features = [total_power, low_freq_power, high_freq_power]
    frequency_features.append(features)
```

```python
# Simple R-wave amplitude feature (peak of the R-wave)
r_wave_amplitudes = [ecg_signal[r_peak] for r_peak in r_peaks]

# Other morphological features would require more sophisticated signal analysis
```

```python
print("Length of time_features:", len(time_features))
print("Length of frequency_features:", len(frequency_features))
print("Length of r_wave_amplitudes:", len(r_wave_amplitudes))
```

```python
# Ensure that the lists have the same number of entries
assert len(time_features) == len(frequency_features) == len(r_wave_amplitudes), "Feature lists must have the same length"

combined_features = [time + freq + [amp] for time, freq, amp in zip(time_features, frequency_features, r_wave_amplitudes)]
```

```python
# Create a dataframe using the combined features
import pandas as pd

feature_columns = ['Mean', 'StdDev', 'TotalPower', 'LowFreqPower', 'HighFreqPower', 'RWaveAmplitude']
feature_df = pd.DataFrame(combined_features, columns=feature_columns)
```
